```python
# 1. Distribution of patients' ages across different genders and ethnicities
plt.figure(figsize=(10, 6))
sns.histplot(df, x='Age', hue='Gender', multiple='stack', kde=True)
plt.title('Distribution of Age by Gender')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Ethnicity', y='Age', hue='Gender', data=df)
plt.title('Age Distribution by Ethnicity and Gender')
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. BMI variation with socioeconomic status and education level
plt.figure(figsize=(10, 6))
sns.boxplot(x='SocioeconomicStatus', y='BMI', hue='EducationLevel', data=df)
plt.title('BMI by Socioeconomic Status and Education Level')
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Correlation between physical activity, diet quality, and BMI
plt.figure(figsize=(10, 6))
sns.pairplot(df[['PhysicalActivity', 'DietQuality', 'BMI']], diag_kind='kde')
plt.title('Correlation Matrix for Physical Activity, Diet Quality, and BMI')
plt.show()
```

```python
# 4. Frequency of hypertension by age groups and gender
df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 30, 50, 70, 100], labels=['<30', '30-50', '50-70', '>70'])
plt.figure(figsize=(10, 6))
sns.countplot(x='AgeGroup', hue='Hypertension', data=df)
plt.title('Hypertension Frequency by Age Group')
plt.show()
```

```python
# 5. Average fasting blood sugar and HbA1c levels by gender and ethnicity
plt.figure(figsize=(10, 6))
sns.barplot(x='Ethnicity', y='FastingBloodSugar', hue='Gender', data=df, errorbar='sd')
plt.title('Average Fasting Blood Sugar by Gender and Ethnicity')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Ethnicity', y='HbA1c', hue='Gender', data=df, errorbar='sd')
plt.title('Average HbA1c by Gender and Ethnicity')
plt.xticks(rotation=45)
plt.show()
```

```python
# 6. Relationship between alcohol consumption and cholesterol levels
plt.figure(figsize=(10, 6))
sns.scatterplot(x='AlcoholConsumption', y='CholesterolTotal', data=df)
plt.title('Alcohol Consumption vs. Total Cholesterol')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='AlcoholConsumption', y='CholesterolLDL', data=df, label='LDL')
sns.lineplot(x='AlcoholConsumption', y='CholesterolHDL', data=df, label='HDL')
sns.lineplot(x='AlcoholConsumption', y='CholesterolTriglycerides', data=df, label='Triglycerides')
plt.title('Alcohol Consumption vs. Cholesterol Levels')
plt.show()
```

```python
# 7. Medication adherence effect on fasting blood sugar and HbA1c
plt.figure(figsize=(10, 6))
sns.boxplot(x='MedicationAdherence', y='FastingBloodSugar', data=df)
plt.title('Fasting Blood Sugar by Medication Adherence')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='MedicationAdherence', y='HbA1c', data=df)
plt.title('HbA1c by Medication Adherence')
plt.show()
```

```python
# 8. Occurrence of prediabetes and family history of diabetes
plt.figure(figsize=(10, 6))
sns.countplot(x='FamilyHistoryDiabetes', hue='PreviousPreDiabetes', data=df)
plt.title('Prediabetes Occurrence by Family History of Diabetes')
plt.show()
```

```python
# 9. Medical checkups frequency vs diagnosis rates of diabetes/prediabetes
plt.figure(figsize=(10, 6))
sns.countplot(x='MedicalCheckupsFrequency', hue='Diagnosis', data=df)
plt.title('Diagnosis by Medical Checkups Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
# 10. Quality of life vs fatigue, physical activity, and sleep quality
plt.figure(figsize=(10, 6))
sns.heatmap(df[['FatigueLevels', 'PhysicalActivity', 'SleepQuality', 'QualityOfLifeScore']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Fatigue, Physical Activity, Sleep Quality, and Quality of Life')
plt.show()
```
