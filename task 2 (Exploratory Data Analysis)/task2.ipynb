```python
data2.columns #detect outliers using IQR 
```

```python
data2.drop(['video_id', 'trending_date','title','channel_title','publish_time','tags','thumbnail_link','comments_disabled','ratings_disabled','video_error_or_removed','description'], axis=1, inplace=True)
data2.columns
```

```python
Q1=data2.quantile(0.25)
Q3=data2.quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

```python
data2=data2[~((data2< (Q1-1.5*IQR) ) | (data2>(Q3+1.5*IQR)) ) .any(axis=1)]
data2
```

```python
data2.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data2['views'], bins=40, kde=True, color='skyblue')
plt.title('Distribution of Views')
plt.xlabel('Views')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='views', y='likes', data=data2, alpha=0.5, color='coral')
plt.title('Scatter Plot of Likes vs. Views')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.show()
```

```python
plt.figure(figsize=(14, 7))
sns.boxplot(data=data2[['likes', 'dislikes', 'comment_count']])
plt.title('Box Plot of Likes, Dislikes, and Comment Count')
plt.xlabel('Metrics')
plt.ylabel('Values')
plt.show()
```

```python
plt.figure(figsize=(10, 10))
sns.pairplot(data2[['views', 'likes', 'dislikes', 'comment_count']])
plt.suptitle('Pair Plot of Views, Likes, Dislikes, and Comment Count', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
correlation_matrix = data2[['views', 'likes', 'dislikes', 'comment_count']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
